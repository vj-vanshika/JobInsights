```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

## Download the chromedriver from link in description
## And give the location of executable here
driver = webdriver.Chrome()

dataframe = pd.DataFrame(columns=["Title", "Location", "Company", "Salary", "Sponsored", "Description"])

for i in range(0, 1, 10):

    ##Step1: Get the page
    driver.get("https://www.simplyhired.com/search?q=python+developer&l=los-angeles%2C+ca")
    driver.implicitly_wait(4)
    wait = WebDriverWait(driver, 10)  # Wait for a maximum of 10 seconds
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

    all_jobs = driver.find_elements(By.CLASS_NAME, "css-0")
    print(all_jobs)
    for job in all_jobs:
        result_html = job.get_attribute('innerHTML')
        soup = BeautifulSoup(result_html, 'html.parser')
        print(soup)

        try:
            title = driver.find_element(By.CSS_SELECTOR, "a.chakra-button.css-1y7j4hn").text.replace('\n', '')
        except:
            title = 'None'
        print(title)

        try:
            location = driver.find_element(By.CLASS_NAME, "css-1t92pv").text
        except:
            location = 'None'
        print(location)

        try:
            company = driver.find_element(By.CSS_SELECTOR, "span[data-testid='companyName']").text.replace("\n", "").strip()
        except:
            company = 'None'
        print(company)

        try:
            salary = driver.find_element(By.CSS_SELECTOR, "p.chakra-text.css-1ejkpji").text.replace("\n", "").strip()
        except:
            salary = 'None'
        print(salary)

        link = driver.find_element(By.CSS_SELECTOR, "a.chakra-button.css-1y7j4hn")
        href_link = link.get_attribute("href")
        print(href_link)

        job_desc = driver.find_element(By.CSS_SELECTOR, "p.chakra-text.css-jhqp7z").text
        print(job_desc)

        dataframe = dataframe.append({'Title': title, 'Location': location, "Company": company, "Salary": salary,
                                      "Description": job_desc, 'Job Link': href_link},
                                     ignore_index=True)
driver.quit()

dataframe.to_csv("trial.csv", index=False)
```
